```python
import random
data['Experience'] = pd.Series(random.choices([range(100)], weights=[1,1], k=len(data)), index=data.index)
data
```

```python
import random


def categorise1(row):
  if row['Rating']>4:
    return random.randrange(70,100)
  return random.randrange(0,70)
```

```python

data['Experience'] = data.apply(lambda row:categorise1(row),axis=1)
```

```python
def categorise(row):
  if row['Rating']>4 or row['Experience']>45:
    return 1
  return 0;
```

```python
data['Target'] = data.apply(lambda row:categorise(row),axis=1)
```

```python
features = ['City', 'Place', 'Rating', 'Target']
x=data[features]
y=data['Target']

print(x)
print(y)
```

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

features = ['Rating','Experience']
x=data[features]
y=data['Target']

print(x)
print(y)

X_train,X_test,Y_train,Y_test = train_test_split(x,y,random_state=10,test_size=0.2,shuffle=True)

dtree=DecisionTreeClassifier()
dtree=dtree.fit(X_train,Y_train)

tree.plot_tree(dtree, feature_names=features)
```

```python
prediction = dtree.predict(X_test.values)
```

```python
from sklearn.metrics import accuracy_score
accuracy_dt = accuracy_score(Y_test,prediction)*100
```
